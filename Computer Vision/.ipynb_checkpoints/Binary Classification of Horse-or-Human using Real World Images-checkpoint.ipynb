

## Creating the Model:

We then add concilutional layers as in the previous example, and flattern the final result to feed into the densely connected layers.

Finally we add the densely connected layers.

Note that because we are facing a two-class classification problem, i.e. binary classification problem, we will end our network with a sigmoid activation, so that the output of our netwrok will be a single scalar between 0 and 1, encoding the probablilty that the current mage is class 1 (as opposed to class 0).

```python
%tensorflow_version 1.x
import tensorflow as tf
```

```python
model = tf.keras.models.Sequential([
                                    tf.keras.layers.Conv2D(16, (3,3), activation= 'relu', input_shape =(300, 300, 3)),
                                    tf.keras.layers.MaxPooling2D(2,2),
                                    tf.keras.layers.Conv2D(32, (3,3), activation = 'relu'),
                                    tf.keras.layers.MaxPooling2D(2,2),
                                    tf.keras.layers.Conv2D(64, (3,3), activation = 'relu'),
                                    tf.keras.layers.MaxPooling2D(2,2),
                                    tf.keras.layers.Conv2D(64, (3,3), activation = 'relu'),
                                    tf.keras.layers.MaxPooling2D(2,2),
                                    tf.keras.layers.Conv2D(64, (3,3), activation = 'relu'),
                                    tf.keras.layers.MaxPooling2D(2,2),
                                    tf.keras.layers.Flatten(),
                                    tf.keras.layers.Dense(512, activation='relu'),
                                    tf.keras.layers.Dense(1, activation = 'sigmoid')
])
```

```python
model.summary()
```

### Compiling the Network:

the output shape column shows how the size of your feature map evolves in each successive layer. The convolution layers reduce the size of the feature maps by a bit due to paddingm and each polling layer halves the dimensions.

Now we will use the **loss = binary_crossentropy** because ts an binary classification problem and our final neuron activation is sigmoid.

Also, we will this time use the **RMS Prop optimizer, with a learning _rate of 0.001.** During trainingm we will want to monitor the classificaiton accuracy.

```python
from tensorflow.keras.optimizers import RMSprop

model.compile(optimizer=RMSprop(lr = 0.001), loss = 'binary_crossentropy',metrics=['acc'])
```

### Using ImageDataGenerator

- Step 1: Normalize the Image data: we will change our pixel values in the image from [0, 255] to [0, 1] range.

- Step 2: In keras this can be done using the keras.preprocessing.image.ImageDataGenerator class using the rescale parameter.

  - The ImageDataGenerator class allows you to instantiate generators of augmented image batches( and their labels) via  .flow(data, labels) or  .flow_from_directory(directory)

- Step 3: These generators can then be used with the keras model methods that accept data generators as inputs:
fit_generator, evaluate_generator, and predict_generator.


```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# All images will be rescaled by 1./255
train_datagen = ImageDataGenerator(rescale = 1./255)

# Flow training images in batches of 128 using train_datagen generator
train_generator = train_datagen.flow_from_directory(
    '/tmp/horse-or-human/',
    target_size = (300,300),
    batch_size = 128,
    # Since we use binary_crossentroly loss, we need binary labels
    class_mode = 'binary'
)
```

### Training: 

```python
history = model.fit_generator(train_generator,
                              steps_per_epoch = 8,
                              epochs = 15,
                              verbose = 1)
```

### Running the Model:

```python
import numpy as np
from google.colab import files
from keras.preprocessing import image
uploaded = files.upload()

for fn in uploaded.keys():

  # Predicting Images
  path = '/content/'+ fn
  img = image.load_img(path, target_size = (300,300))
  x = image.img_to_array(img)
  x = np.expand_dims(x, axis= 0)

  images = np.vstack([x])
  classes = model.predict(images, batch_size = 10)
  print(classes[0])
  if classes[0]>0.5:
    print(fn + ' is a human')
  else:
    print(fn + ' is a horse')
```

# Important to Note:

Although I used the evaluate_generator for the validation generator, I can simple take this step and merge it with the training step of the model. When we train the model, there are some additional parameters that can be passed and after every epoch gives us the validation acc and details as well.

```python
history2 = model.fit_generator(
    train_generator,
    steps_per_epoch = 8,
    epochs = 15,
    verbose = 1,
    validation_data = validation_generator,
    validation_steps = 8
)
```
